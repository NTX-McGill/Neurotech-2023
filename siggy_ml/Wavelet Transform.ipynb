```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import cwt, ricker, daub
from erp_classifier import classifier
```

```python
X_train = np.load('Cz_dataset/X_train.npy')
X_test = np.load('Cz_dataset/X_test.npy')
y_train = np.load('Cz_dataset/y_train.npy')
y_test = np.load('Cz_dataset/y_test.npy')
```

```python
0.6*256
```

```python
clf = classifier()
```

```python
widths = np.arange(1, 50)
for i in range(10):
    plt.figure(figsize = (20,6))
    plt.imshow(cwt(clf.preprocess(X_train[i], upper = 40), ricker, widths))
    plt.title("label = " + str(y_train[i]))
    plt.show()
```

```python
widths = np.arange(1, 40)
for i in range(10):
    plt.figure(figsize = (20,6))
    plt.imshow(cwt(clf.preprocess(X_train[i][int(0.5*256):int(0.7*256)], upper = 40), ricker, widths))
    plt.title("label = " + str(y_train[i]))
    plt.show()
```

```python
X_hit = [X_train[i] for i in range(len(X_train)) if y_train[i] == 1]
X_miss = [X_train[i] for i in range(len(X_train)) if y_train[i] == 0]
```

```python
plt.plot(np.mean(clf.preprocess(X_hit), axis = 0) - np.mean(clf.preprocess(X_miss), axis = 0), label = 'n400')
# plt.plot(np.mean(clf.preprocess(X_miss), axis = 0), label = 'no n400')
plt.legend()
```
